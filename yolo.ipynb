욜로 이미지 인식

```python
# 1. 필수 라이브러리 설치 및 드라이브 마운트
!pip install ultralytics -q

from ultralytics import YOLO
from google.colab import drive
from google.colab.patches import cv2_imshow
import cv2
import os

# 구글 드라이브 연결 (실행 후 나타나는 팝업에서 승인 필요)
drive.mount('/content/drive')

# 2. 경로 설정 및 파일 확인
# 사용자 폴더 경로 (제공해주신 폴더 ID 기반 경로는 마운트 후 '내 드라이브' 내 실제 경로로 수정 필요)
# 아래 경로는 일반적인 '내 드라이브' 루트 기준 예시입니다.
folder_path = '/content/drive/MyDrive/'
image_name = 'img2.jpg'
input_path = os.path.join(folder_path, image_name)

# 3. 모델 로드 (yolov11s.pt)
model = YOLO('yolov11s.pt')

# 4. 이미지 분석 및 결과 도출
if os.path.exists(input_path):
    results = model.predict(source=input_path, classes=[0], conf=0.25, save=False)
    person_count = len(results[0].boxes)

    # 5. 요청 형식에 따른 결과 출력
    print("\n" + "="*30)
    print(f" 이미지의 사람수 : {person_count}")
    print("="*30 + "\n")

    # 결과 이미지 시각화
    annotated_img = results[0].plot()
    cv2_imshow(annotated_img)
else:
    print(f"[오류] 경로에 {image_name} 파일이 존재하지 않습니다. 경로를 확인해주세요.")
```
